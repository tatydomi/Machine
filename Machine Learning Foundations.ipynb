```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
from sklearn.cluster import KMeans, DBSCAN
from sklearn.datasets import make_blobs, make_moons, make_circles, load_iris
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, calinski_harabasz_score
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import train_test_split
```

# Clusterização

## Aplicando K-means a exemplo didático

```python
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.7, random_state=0)

plt.scatter(X[:,0], X[:,1])

plt.show()
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', n_init=10, max_iter=300, tol=1**-4, random_state=0)

pred_y = kmeans.fit_predict(X)

plt.scatter(X[:,0], X[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.scatter(X[:, 0], X[:, 1], c=pred_y)

score_silhouette = silhouette_score(X, pred_y)
score_ch = calinski_harabasz_score(X, pred_y)

print(f"score silhueta {score_silhouette:.2f}, Calinski and Harabasz {score_ch:.2f}")

plt.show()
```

```python
dbscan = DBSCAN(eps=1, min_samples=10)

pred_y = dbscan.fit_predict(X)

plt.scatter(X[:,0], X[:,1])
plt.scatter(X[:, 0], X[:, 1], c=pred_y)

score_silhouette = silhouette_score(X, pred_y)
score_ch = calinski_harabasz_score(X, pred_y)

print(f"score silhueta {score_silhouette:.2f}, Calinski and Harabasz {score_ch:.2f}")

plt.show()
```

```python
gmm = GaussianMixture(n_components=4, covariance_type="full", random_state=42)

pred_y = gmm.fit_predict(X)

plt.scatter(X[:,0], X[:,1])
plt.scatter(X[:, 0], X[:, 1], c=pred_y)

score_silhouette = silhouette_score(X, pred_y)

score_silhouette = silhouette_score(X, pred_y)
score_ch = calinski_harabasz_score(X, pred_y)

print(f"score silhueta {score_silhouette:.2f}, Calinski and Harabasz {score_ch:.2f}")

plt.show()
```

# Teste de Hipotese

```python
mean, std, num_elements = 0, 0.1, 100_000
normal_dist = np.random.normal(mean, std, num_elements)

mean_2, std_2, num_elements_2 = 2, 0.1, 100_000
normal_dist_2 = np.random.normal(mean_2, std_2, num_elements_2)

plt.hist(normal_dist, 300, density=True)
plt.hist(normal_dist_2, 300, density=True)

plt.show()
```

**1-O que acontece se mudar std_2 para 0.4?**
**R:** O espaçamento diminui entre as hipoteses.

```python
mean, std, num_elements = 0, 0.4, 100_000
normal_dist = np.random.normal(mean, std, num_elements)

mean_2, std_2, num_elements_2 = 2, 0.4, 100_000
normal_dist_2 = np.random.normal(mean_2, std_2, num_elements_2)

plt.hist(normal_dist, 300, density=True)
plt.hist(normal_dist_2, 300, density=True)

plt.show()
```

**2-O que acontece com o teste caso mude mean_2 para 3?**
**R:** Da um espaçamento entre o um e o dois maior no eixo x

```python
mean, std, num_elements = 0, 0.4, 100_000
normal_dist = np.random.normal(mean, std, num_elements)

mean_2, std_2, num_elements_2 = 3, 0.4, 100_000
normal_dist_2 = np.random.normal(mean_2, std_2, num_elements_2)

plt.hist(normal_dist, 300, density=True)
plt.hist(normal_dist_2, 300, density=True)

plt.show()
```

**3-Aplique o teste de hipotese para o caso da iris (ver aula 2), a largura da sepala e comprimento são diferentes entre as classes?**
**R:**

```python
stats_results = stats.ttest_ind(normal_dist, normal_dist_2, equal_var=False)

print(f"resultado t test {stats_results.statistic}, p valor {stats_results.pvalue}")
```
